В данном наборе данных переменные можно разделить на три типа по шкале измерения. Номинальные переменные, которые представляют собой категории без внутреннего порядка, включают школу, пол, тип домашнего адреса, размер семьи, статус совместного проживания родителей, работу матери, работу отца, причину выбора школы, опекуна, а также ряд бинарных признаков: наличие дополнительной образовательной поддержки, семейной образовательной поддержки, дополнительных платных занятий, внеучебных занятий, факт посещения детского сада, желание получить высшее образование, доступ в интернет и наличие романтических отношений. К порядковым переменным, значения которых можно ранжировать, относятся образование матери, образование отца, время в пути до школы, время на учёбу, качество семейных отношений, объём свободного времени, время с друзьями, уровень потребления алкоголя в будни и выходные, а также текущее состояние здоровья. Числовые переменные, представленные количественными значениями, — это возраст, количество неудач по предметам, количество пропущенных занятий и оценки за первый, второй период и итоговая оценка.

```python
nominal_variables = ["school", "sex", "address", "famsize", "Pstatus", "Mjob", "Fjob", "reason", "guardian", "schoolsup", "famsup",  "paid",    "activities", "nursery", "higher", "internet", "romantic"]

for column in nominal_variables:
	print(f'{column}: {df[column].nunique()} уникальных значений')
```

```python
ordinal_variables = ["Medu", "Fedu", "traveltime", "studytime", "famrel", "freetime", "goout", "Dalc", "Walc", "health"]

for column in ordinal_variables:
	print(f'{column}: {df[column].nunique()} уникальных значений')
```

```python
numeric_variables = ["age", "failures", "absences", "G1", "G2", "G3"]

for column in numeric_variables:
	print(f'{column}: {df[column].nunique()} уникальных значений')
```

Выведем гистограмму для переменной absences, которая хранит количество пропусков в школе среди студентов:

```python
sns.histplot(df['absences'], kde=True)
plt.title('Распределение absences')
plt.xlabel('absences')
plt.ylabel('Частота')
plt.show()
```

Данное распределение нельзя считать нормальным из-за отсутствия симметрии и формы колокола. Для него характерна высокая концентрация значений в диапазоне от 0 до 10.

Теперь выведем другие пять числовых переменных аналогичным образом:

```python
num_cols = df.columns.intersection(numeric_variables)

fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

Ни одна переменная не является нормально распределённой из-за отсутствия симметричности в распределении и много значений превышают остальные.

Выведем столбчатую диграмму по категориальной переменной health, которая показывает уровень здоровья студентов курса:

```python
sns.countplot(x='health', data=df)
plt.title('Распределение health')
plt.xlabel('health')
plt.ylabel('Количество')
plt.show()
```

Значения концентрируются при health равном 3 и 5.

Выведем также гистограммы для 4-х других переменных:

```python
num_cols = df.columns.intersection(['Medu', 'Fedu', 'Dalc', 'Walc'])
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.countplot(x=col, data=df, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Количество')

plt.tight_layout()
plt.show()
```

Выведем боксплот для переменной age, которая показывает возраст студентов:

```python
sns.boxplot(x=df['age'])
plt.title('Boxplot age')
plt.xlabel('age')
plt.show()
```

На графике можно увидеть выброс при значении равном 22

Построим корреляционную матрицу для всех числовых переменных и визуализируем её с помощью тепловой карты:

```python
correlation_matrix = df[numeric_variables].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

Анализ корреляций между переменными выявил три характерные группы. Пары age/G1, age/G2, age/G3 и age/failures демонстрируют практически полное отсутствие линейной связи. В то же время, наблюдается слабая отрицательная корреляция между количеством неудач (failures) и академическими оценками (G1, G2, G3), что указывает на незначительное снижение успеваемости при увеличении числа провалов. Наиболее сильная положительная корреляция зафиксирована между самими оценками (G1/G2, G1/G3, G2/G3), что свидетельствует об их высокой взаимосвязанности и устойчивости академических результатов в течение периода.
